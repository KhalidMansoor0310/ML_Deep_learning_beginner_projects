```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/car data.csv')
df.head()
```

```python
inputs = df.drop(['Car_Name','Owner','Seller_Type','Present_Price'], axis=1)
target = df.Selling_Price
```

```python
target
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
inputs['Transmission_new']=le.fit_transform(df['Transmission'])
inputs['fuelType_new']=le.fit_transform(df['Fuel_Type'])

inputs
```

```python
X = inputs.drop(['Transmission','Fuel_Type','Selling_Price'],axis='columns')
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,target)
```

```python
model.score(X,target)
```
